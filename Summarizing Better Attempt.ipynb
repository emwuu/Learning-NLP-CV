```python
import urllib2
from bs4 import BeautifulSoup
```

```python
def scrape(url):
    page=urllib2.urlopen(url).read().decode('utf8', 'ignore')
    soup = BeautifulSoup(page, "lxml")
    return soup
```

```python
page = scrape('https://www.nytimes.com/2019/02/02/world/asia/japan-working-mothers.html?module=inline')
print(page)
```

```python
tags = page.find_all('div', attrs={'StoryBodyCompanionColumn'})
```

```python
text = ' '.join(map(lambda t: t.text, tags))
```

```python
text = text.encode('ascii', errors='replace').replace("?", " ")
text = text.replace(".", ". ")
print(text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(text)
```

```python
print(sents)
```

```python
word_sent = word_tokenize(text.lower())
```

```python
_stopwords=set(stopwords.words('english') + list(punctuation) + ["ms.", "nishimasa", "said"])
```

```python
word_sent =[word for word in word_sent if word not in  _stopwords]
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(word_sent)
```

```python
freq
```

```python
from heapq import nlargest
nlargest(12, freq, key=freq.get)
```

```python
from collections import defaultdict
```

```python
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
```

```python
sents_idx = nlargest(3, ranking, key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
